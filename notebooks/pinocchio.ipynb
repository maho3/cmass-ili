```python
def compute_pk(delta, boxsize=1000, MAS='CIC'):

    Pk = PKL.Pk(delta.astype(np.float32), boxsize, axis=0, MAS=MAS, threads=1, verbose=False)
    k3D      = Pk.k3D
    Pk3D     = Pk.Pk[:,0]
    
    return k3D, Pk3D


def compute_crossk(delta1, delta2, boxsize=1000, MAS=['CIC','CIC']):
    
    Pk = PKL.XPk([delta1.astype(np.float32),delta2.astype(np.float32)], boxsize, 0, MAS=MAS, threads=1)
    
    k      = Pk.k3D
    Pk0_1  = Pk.Pk[:,0,0]  #monopole of field 1
    Pk0_2  = Pk.Pk[:,0,1]  #monopole of field 2
    Pk0_X  = Pk.XPk[:,0,0] #monopole of 1-2 cross P(k)
    
    r = Pk0_X / np.sqrt(Pk0_1 * Pk0_2)
    
    return k, Pk0_1, Pk0_2, r


N = 128
L = 1000
MAS = 'CIC'

delta_pinocchio = np.zeros((N,N,N), dtype=np.float32)
MASL.MA(data['positions'].astype(np.float32), delta_pinocchio, L, MAS, verbose=True)
delta_pinocchio = np.transpose(delta_pinocchio, (2,1,0))

delta_pinocchio = np.load(pjoin(dirname, 'rho.npy'))

delta_quijote = np.load(f'/home/mattho/data/quijote/density_field/latin_hypercube/3/df_m_{N}_z=0.npy')

k, Pk0_pinocchio, Pk0_quijote, r_pinocchio = compute_crossk(delta_pinocchio, delta_quijote)

fig, axs = plt.subplots(1, 3, figsize=(15,4))
axs[0].loglog(k, Pk0_pinocchio, label='Pinocchio')
axs[0].loglog(k, Pk0_quijote, label='Quijote')
axs[0].set_ylabel(r'$P(k)$')
axs[1].semilogx(k, Pk0_pinocchio/ Pk0_quijote, label='Pinocchio')
axs[1].axhline(1, color='k')
axs[1].set_ylabel('P(k) / Quijote')
axs[2].semilogx(k, r_pinocchio, label='Pinocchio')
axs[2].axhline(1, color='k')
axs[2].set_ylabel('Cross Correlation with Quijote')
for ax in axs:
    ax.set_xlabel(r'$k$')
    ax.legend()
fig.tight_layout()
fig.savefig('../images/pinocchio_pk.png', facecolor='white', bbox_inches='tight')

fig, axs = plt.subplots(1, 2, figsize=(10,4))

vmin = np.amin(np.log10(2 + delta_quijote[N//2]))
vmax = np.amax(np.log10(2 + delta_quijote[N//2]))
x = np.linspace(0, L, N)

pc = axs[0].pcolor(x, x, np.log10(2 + delta_quijote[N//2]), vmin=vmin, vmax=vmax)
cb = fig.colorbar(pc, ax=axs[0])
cb.set_label(r'$\log_{10}(2 + \delta)$')
axs[0].set_title('Quijote')

pc = axs[1].pcolor(x, x, np.log10(2 + delta_pinocchio[N//2]), vmin=vmin, vmax=vmax)
cb = fig.colorbar(pc, ax=axs[1])
cb.set_label(r'$\log_{10}(2 + \delta)$')
axs[1].set_title('Pinocchio')

for ax in axs:
    ax.set_aspect('equal')
    ax.set_xlabel(r'$x \ [{\rm Mpc}/h]$')
    ax.set_ylabel(r'$y \ [{\rm Mpc}/h]$')
fig.tight_layout()
fig.savefig('../images/pinocchio_field.png', facecolor='white', bbox_inches='tight')
```

# Analyse halo catalogue

```python
#    0) group ID
#    1) group mass (Msun/h)
# 2- 4) initial position (Mpc/h)
# 5- 7) final position (Mpc/h)
# 8-10) velocity (km/s)
#   11) number of particles
sim_num = 3
dirname = '/data101/bartlett/ili/cmass/quijote_z0/pinocchio/L1000-N128/3/'
filename = pjoin(dirname, 'pinocchio.0.0000.pinocchio-L1000-N128-3.catalog.out')

halo_dir = '/home/mattho/data/quijote/Halos/latin_hypercube/'

# Load pinocchio halos
# mass_pinocchio = np.log10(np.loadtxt(filename,unpack=False,usecols=(1,)))
# pos_pinocchio = np.loadtxt(filename,unpack=False,usecols=(7,6,5))
# print(mass_pinocchio.min(), mass_pinocchio.max())
mass_pinocchio = np.load(pjoin(dirname, 'halo_mass.npy'))
pos_pinocchio = np.load(pjoin(dirname, 'halo_pos.npy'))
vel_pinocchio = np.load(pjoin(dirname, 'halo_vel.npy'))

# Load Quijote halos
snapdir = pjoin(halo_dir, str(sim_num))
snapnum = 4; redshift = 0
FoF = readfof.FoF_catalog(snapdir, snapnum, long_ids=False,
                          swap=False, SFR=False, read_IDs=False)
pos_quijote = FoF.GroupPos/1e3             #Halo positions in Mpc/h
mass_quijote  = np.log10(FoF.GroupMass*1e10) #Halo masses in Msun/h
vel_quijote = FoF.GroupVel * (1.0 + redshift)  # Halo velocities in km/s
print(mass_quijote.min(), mass_quijote.max())

mmin = 14.8; mmax = 15.8

# n(M) 
fig, ax = plt.subplots(figsize=(5,4))
bins = np.linspace(mmin, mmax, 30)
hist_kwargs = dict(density=False, histtype='step', bins=bins)
ax.hist(mass_pinocchio[mass_pinocchio>mmin], label='Pinocchio', **hist_kwargs)
ax.hist(mass_quijote[mass_quijote>mmin], label='Quijote', **hist_kwargs)
ax.legend()
ax.set_yscale('log')
ax.set_xlabel(r'$\log_{10} (M \ / \ {\rm M}_{\odot})$')
ax.set_ylabel('Number of Halos')
fig.tight_layout()
fig.savefig('../images/pinocchio_hmf.png', facecolor='white', bbox_inches='tight')

hf_pinocchio = np.zeros((N,N,N), dtype=np.float32)
MASL.MA(
    pos_pinocchio[(mass_pinocchio > mmin) & (mass_pinocchio < mmax)].astype(np.float32), 
    hf_pinocchio, L, MAS, verbose=True)
hf_quijote = np.zeros((N,N,N), dtype=np.float32)
MASL.MA(
    pos_quijote[(mass_quijote > mmin) & (mass_quijote < mmax)].astype(np.float32), 
    hf_quijote, L, MAS, verbose=True)

k, Pk0_pinocchio, Pk0_quijote, r_pinocchio = compute_crossk(hf_pinocchio, hf_quijote)

# Halo counts field
fig, axs = plt.subplots(1, 3, figsize=(15,4))
axs[0].loglog(k, Pk0_pinocchio, label='Pinocchio')
axs[0].loglog(k, Pk0_quijote, label='Quijote')
axs[0].set_ylabel(r'$P(k)$')
axs[1].semilogx(k, Pk0_pinocchio/ Pk0_quijote, label='Pinocchio')
axs[1].axhline(1, color='k')
axs[1].set_ylabel('P(k) / Quijote')
axs[2].semilogx(k, r_pinocchio, label='Pinocchio')
axs[2].axhline(1, color='k')
axs[2].set_ylabel('Cross Correlation with Quijote')
for ax in axs:
    ax.set_xlabel(r'$k$')
    ax.legend()
axs[1].set_title(r'Halo Counts Field: $\log_{10} (M_{\rm min} \ / \ {\rm M}_{\odot}) = %.2f$, $\log_{10} (M_{\rm max} \ / \ {\rm M}_{\odot}) = %.2f$'%(mmin, mmax))
fig.tight_layout()
fig.savefig('../images/pinocchio_halo_pk.png', facecolor='white', bbox_inches='tight')

# Velocity distributions
fig, ax = plt.subplots(figsize=(5,4))
vp = np.sqrt(np.sum(vel_pinocchio[(mass_pinocchio > mmin) & (mass_pinocchio < mmax)]**2, axis=1))
vq = np.sqrt(np.sum(vel_quijote[(mass_quijote > mmin) & (mass_quijote < mmax)]**2, axis=1))
bins = np.linspace(0, max(vp.max(), vq.max()), 30)
hist_kwargs = dict(density=False, histtype='step', bins=bins)
ax.hist(vp, label='Pinocchio', **hist_kwargs)
ax.hist(vq, label='Quijote', **hist_kwargs)
ax.legend()
ax.set_xlabel(r'$|v| \ / \ {\rm km \, s^{-1}}$')
ax.set_ylabel('Number of halos')
ax.set_ylim(0, None)
fig.tight_layout()
fig.savefig('../images/pinocchio_vel_hist.png', facecolor='white', bbox_inches='tight')

# Check transposes
fig, axs = plt.subplots(1, 3, figsize=(15,4), sharex=True, sharey=True)
for i in range(3):
    hf_pinocchio = np.zeros((N,N,N), dtype=np.float32)
    m = (mass_pinocchio > mmin) & (mass_pinocchio < mmax)
    MASL.MA( 
        pos_pinocchio[m].astype(np.float32),
        hf_pinocchio, L, MAS, 
        W=vel_pinocchio[m,i].astype(np.float32), verbose=False)
    hf_quijote = np.zeros((N,N,N), dtype=np.float32)
    m = (mass_quijote > mmin) & (mass_quijote < mmax)
    MASL.MA(
        pos_quijote[m].astype(np.float32),
        hf_quijote, L, MAS, 
        W=vel_quijote[m,i].astype(np.float32), verbose=False)
    k, Pk0_pinocchio, Pk0_quijote, r_pinocchio = compute_crossk(hf_pinocchio, hf_quijote)
    axs[i].semilogx(k, r_pinocchio, label='Pinocchio')
    axs[i].axhline(1, color='k')
    coord = ['x', 'y', 'z']
    axs[i].set_title(r'$v_{%s}$'%coord[i])
    axs[i].set_xlabel(r'$k$')
    axs[i].legend(loc='lower left')
axs[0].set_ylabel(r'Cross Correlation with Quijote')
fig.align_labels()
fig.tight_layout()
fig.savefig('../images/pinocchio_vel_rk.png', facecolor='white', bbox_inches='tight')
```
